```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# -------------------------------
# Step 1: Generate Fake Accident Data
# -------------------------------
np.random.seed(42)
n = 500

df = pd.DataFrame({
    'accident_id': range(1, n+1),
    'time_of_day': np.random.choice(['Morning', 'Afternoon', 'Evening', 'Night'], size=n),
    'weather': np.random.choice(['Clear', 'Rain', 'Fog', 'Snow', 'Cloudy'], size=n, p=[0.5, 0.2, 0.1, 0.05, 0.15]),
    'road_condition': np.random.choice(['Dry', 'Wet', 'Icy', 'Snowy'], size=n, p=[0.6, 0.3, 0.05, 0.05]),
    'severity': np.random.choice(['Minor', 'Moderate', 'Severe'], size=n, p=[0.6, 0.3, 0.1]),
    'location': np.random.choice(['Downtown', 'Highway', 'Suburb', 'Intersection'], size=n),
    'vehicles_involved': np.random.randint(1, 5, size=n)
})

print("✅ Sample Accident Data:")
print(df.head())

# -------------------------------
# Step 2: Data Summary
# -------------------------------
print("\n✅ Summary Statistics:")
print(df.describe(include='all'))

print("\n✅ Missing Values:")
print(df.isnull().sum())

# -------------------------------
# Step 3: Visualizations
# -------------------------------

# Accident count by time of day
sns.countplot(data=df, x='time_of_day', order=['Morning', 'Afternoon', 'Evening', 'Night'])
plt.title("Accidents by Time of Day")
plt.ylabel("Number of Accidents")
plt.show()

# Severity by weather
sns.countplot(data=df, x='weather', hue='severity')
plt.title("Accident Severity by Weather")
plt.ylabel("Count")
plt.xticks(rotation=30)
plt.show()

# Road condition vs severity
sns.countplot(data=df, x='road_condition', hue='severity')
plt.title("Severity by Road Condition")
plt.ylabel("Number of Accidents")
plt.show()

# Location hotspots
sns.countplot(data=df, x='location')
plt.title("Accident Hotspots by Location")
plt.ylabel("Accident Count")
plt.show()

# Vehicles involved distribution
sns.histplot(df['vehicles_involved'], bins=4, kde=False)
plt.title("Distribution of Vehicles Involved")
plt.xlabel("Number of Vehicles")
plt.ylabel("Frequency")
plt.show()
```
